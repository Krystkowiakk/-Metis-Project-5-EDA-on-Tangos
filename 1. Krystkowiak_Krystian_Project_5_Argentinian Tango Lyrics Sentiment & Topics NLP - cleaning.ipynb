```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('project.csv')
```

```python
df.info()
```

## CLEANING

```python
df.drop(columns=[' BPM', ' POPULARITY', 'AUTHOR (2)', 'DIRECTOR', 'SOLOIST', 'URL'], inplace=True)
df
```

```python
#getting rid of missing artist - small numbers of rows and instrumental songs - about 50% of rows
df.dropna(inplace=True)
df.LYRICS = df.LYRICS[df.LYRICS != 'EL RECODO TANGO\n\t\t\tDesign Web: Gregory DIAZ']
df.LYRICS[4]
```

```python
#extracting year as only year is my interest and not all songs have detailed date
df["DATE"] = df.DATE.apply(lambda x : str(x)[:4])
df
```

```python
df.DATE
```

```python
#choosing dates between 1920 and 1960
df["DATE"] = df["DATE"].astype(int)
df = df[(df.DATE >= 1920) & (df.DATE <= 1960)]
```

```python
#type extraction
df['TYPE'] = df['TITLE'].str.split().str[-1]
df['TITLE'] = df['TITLE'].str.split().str[:-1]
df['TYPE'] = df["TYPE"].str.strip("()")
df["TYPE"].value_counts()
```

```python
df = df[(df["TYPE"] == "TANGO") | (df["TYPE"] == "VALS") | (df["TYPE"] == "MILONGA")]
df["TYPE"].value_counts()
```

```python
#dropping instrumental song and couple without singer information 
df = df[(df.SINGER != "Instrumental") & (df.SINGER != "unknown")]
```

```python
#dropping duplicates, they are not very important statistictly
df.LYRICS = df.LYRICS.drop_duplicates()
df = df.dropna()
df
```

```python
df["LYRICS"] = df["LYRICS"].str.replace("\n", ' ')
df["LYRICS"] = df["LYRICS"].str.replace("!", ' ')
df["LYRICS"] = df["LYRICS"].str.replace("¡", ' ')
```

```python
df['WORDS'] = df.LYRICS.str.split().str.len()
```

```python
df['WORDS'].sum()
```

```python
#that is dataset to be analised
df.reset_index(inplace=True)
df.to_pickle("./df_project 5.pkl")
df
```
